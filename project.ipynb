```python
# Task 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 2: Load the dataset and display first rows
df = pd.read_csv("/content/SuperMarket Analysis.csv")
df.head()
```

```python
# Task 3: Check dataset information (columns, data types, missing values)
df.info()
```

```python
# Task 4: Data Cleaning
print(df.isnull().sum())      # فحص القيم الناقصة
df = df.dropna()
```

```python
# Task 5: Generate basic statistical summary for numerical columns
df.describe()
```

```python
# Task 6: Basic numerical analysis using NumPy
import numpy as np

average_sales = np.mean(df['Sales'])
median_sales = np.median(df['Sales'])
max_sales = np.max(df['Sales'])
min_sales = np.min(df['Sales'])

print("Average Sales:", average_sales)
print("Median Sales:", median_sales)
print("Maximum Sales:", max_sales)
print("Minimum Sales:", min_sales)
```

```python
# Task 7: Plot distribution of Sales values
df['Sales'].hist()
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Task 8 : Revenue Analysis
total_revenue = df['Sales'].sum()
average_revenue = df['Sales'].mean()

print("Total Revenue:", total_revenue)
print("Average Revenue:", average_revenue)

# Task : Revenue Analysis by Product Line
revenue_by_product = df.groupby('Product line')['Sales'].sum()

print("Revenue by Product Line:")
print(revenue_by_product)

revenue_by_product.plot(kind='bar')
plt.title('Total Revenue by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Revenue')
plt.show()

```

```python
# Task 9: Analyze total sales by branch
df.groupby('Branch')['Sales'].sum()
# Plot total sales by branch
df.groupby('Branch')['Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Task 10: Compare average sales by customer type
df.groupby('Customer type')['Sales'].mean()
# Plot average sales by customer type
df.groupby('Customer type')['Sales'].mean().plot(kind='bar')
plt.title('Average Sales by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Average Sales')
plt.show()
```

```python
# Task 11: Analyze total sales by gender
df.groupby('Gender')['Sales'].sum()
# Plot total sales by gender
df.groupby('Gender')['Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Task 12: Count transactions by payment method
df['Payment'].value_counts()
# Plot payment method usage
df['Payment'].value_counts().plot(kind='bar')
plt.title('Payment Method Usage')
plt.xlabel('Payment Method')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
# Task 13: Calculate average rating for each product line
df.groupby('Product line')['Rating'].mean().sort_values(ascending=False)
# Plot average rating by product line
df.groupby('Product line')['Rating'].mean().sort_values().plot(kind='barh')
plt.title('Average Rating by Product Line')
plt.xlabel('Average Rating')
plt.ylabel('Product Line')
plt.show()
```

```python
# Task 14: Analyze relationship between unit price and quantity
plt.scatter(df['Unit price'], df['Quantity'])
plt.title('Unit Price vs Quantity Purchased')
plt.xlabel('Unit Price')
plt.ylabel('Quantity')
plt.show()
```

```python
# Task 15: Analyze relationship between sales and customer rating
plt.scatter(df['Sales'], df['Rating'])
plt.title('Sales vs Rating')
plt.xlabel('Sales')
plt.ylabel('Rating')
plt.show()
```

```python
# Task 16: Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Task 17: Analyze total sales over time
df.groupby('Date')['Sales'].sum().plot(figsize=(10,5))
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Task 18: Analyze average rating over time
df.groupby('Date')['Rating'].mean().plot(figsize=(10,5))
plt.title('Average Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Task 19: Calculate correlation between numerical variables
corr = df.select_dtypes(include='number').corr()
corr
# Plot correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Task 20: Analyze gross income distribution by product line
df.boxplot(column='gross income', by='Product line', rot=45)
plt.title('Gross Income by Product Line')
plt.suptitle('')
plt.xlabel('Product Line')
plt.ylabel('Gross Income')
plt.show()
```
